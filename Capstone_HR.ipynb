```python
import pandas as pd
import numpy as np
import plotly.graph_objects as go
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score

# Importing classifiers
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from sklearn.svm import SVC
from sklearn.neighbors import KNeighborsClassifier
from xgboost import XGBClassifier
from sklearn.naive_bayes import GaussianNB
from lightgbm import LGBMClassifier

# Splitting the dataset
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=42)

# Dictionary to store models
models = {
    "Logistic Regression": LogisticRegression(),
    "Random Forest": RandomForestClassifier(n_estimators=100, random_state=42),
    "SVM": SVC(kernel='rbf'),
    "KNN": KNeighborsClassifier(n_neighbors=5),
    "XGBoost": XGBClassifier(n_estimators=100, learning_rate=0.1, max_depth=5, random_state=42),
    "Naive Bayes": GaussianNB(),
    "LightGBM": LGBMClassifier(n_estimators=100, learning_rate=0.1, max_depth=5, random_state=42)
}

# Lists to store metrics
accuracy_scores = []
precision_scores = []
recall_scores = []
f1_scores = []
model_names = []

# Training and evaluating each model
for name, model in models.items():
    model.fit(X_train, y_train)
    y_pred = model.predict(X_test)

    # Storing metrics
    accuracy_scores.append(round(accuracy_score(y_test, y_pred), 2))
    precision_scores.append(round(precision_score(y_test, y_pred), 2))
    recall_scores.append(round(recall_score(y_test, y_pred), 2))
    f1_scores.append(round(f1_score(y_test, y_pred), 2))
    model_names.append(name)

# Creating a DataFrame for visualization
metrics_df = pd.DataFrame({
    "Model": model_names,
    "Accuracy": accuracy_scores,
    "Precision": precision_scores,
    "Recall": recall_scores,
    "F1 Score": f1_scores
})

# Sankey Chart Preparation
labels = model_names + ["Accuracy", "Precision", "Recall", "F1 Score"]
sources = []
targets = []
values = []

# Connecting models to metrics
for i, model in enumerate(model_names):
    sources += [i] * 4  # Each model connects to all 4 metrics
    targets += [len(model_names) + j for j in range(4)]  # Targeting metric indices
    values += [accuracy_scores[i], precision_scores[i], recall_scores[i], f1_scores[i]]  # Corresponding metric values

# Create Sankey Diagram
fig = go.Figure(go.Sankey(
    node=dict(
        pad=15, thickness=20,
        line=dict(color="black", width=0.5),
        label=labels,
        color=["blue"] * len(model_names) + ["red", "green", "orange", "purple"]  # Models in blue, metrics in distinct colors
    ),
    link=dict(
        source=sources,
        target=targets,
        value=values
    )
))

fig.update_layout(title_text="Sankey Diagram: Model Performance Metrics", font_size=12)
fig.show()
```
